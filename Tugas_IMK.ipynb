```python
df = df.drop(columns= ["1",	"2",	"3",	"4",	"5",	"6","9"])
```

```python
Total = df.isnull().sum().sort_values(ascending=False)          

Percent = (df.isnull().sum()*100/df.isnull().count()).sort_values(ascending=False)   

missing_data = pd.concat([Total, Percent], axis = 1, keys = ['Total', 'Percentage of Missing Values'])    
missing_data
```

```python
cols = df.columns

for col in cols:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
palette_color = sns.color_palette('crest')

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:d}".format(absolute)

for i, col in enumerate(cols):
  
  plt.figure(figsize=(20,10))
  print(i)
  plt.pie(df[col].value_counts(), labels=df[col].value_counts().keys(), startangle = 90,
          colors=palette_color, autopct=lambda pct: func(pct, df[col].value_counts()),
          wedgeprops = {'linewidth': 6}, center = (0.1,0.1),
        rotatelabels = 'true')
  plt.savefig('/content/drive/MyDrive/chart_imk/nomer ke-{}.jpg'.format(i),bbox_inches='tight', dpi=300)
```
